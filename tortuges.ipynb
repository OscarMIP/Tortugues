```python
import numpy as np
import random

def generar_pregunta(dificultat):
    
    if dificultat == 1:
        num1 = random.randint(0,9)
        num2 = random.randint(0,9)
        return f"{num1} + {num2}", num1 + num2
    elif dificultat == 2:
        num1 = random.randint(0,90)
        num2 = random.randint(0,90)
        return f"{num1} + {num2}", num1 + num2
    elif dificultat == 3:
        num1 = random.randint(0,9)
        num2 = random.randint(0,9)
        return f"{num1} * {num2}", num1 * num2

def mover_jugador(jugador, casillas):
    nueva_posicion = max(1, tauler[jugador, 0] + casillas)  
    tauler[jugador, 0] = nueva_posicion

def imprimir_tablero():
    print("Tablero actualizado:")
    for i, jugador in enumerate(noms_jugadors, start=1):
        print(f"Jugador {i}: {'-' * (tauler[i-1, 0]-1) + 'X' + '-' * (tamany_tauler - tauler[i-1, 0])}")



print("Benvinguts a la carrera de tortuguetes matemàtiques! A continuació començarà el joc.")

n_jugadors = int(input("Quants jugadors jugaran? "))
noms_jugadors = [input(f"Nom del Jugador {i+1}: ") for i in range(n_jugadors)]

tamany_tauler = random.randint(10, 40)
tauler = np.ones((n_jugadors, 1), dtype=int)

print(f"El joc està a punt de començar amb els següents jugadors:")
for i, nom in enumerate(noms_jugadors, start=1):
    print(f"Jugador {i}: {nom}")

print(f"El tamany del tauler és de {tamany_tauler} caselles.")
print("Tablero inicial:")
imprimir_tablero()

meta_alcanzada = False

while not meta_alcanzada:
    for i, jugador in enumerate(noms_jugadors):
        dificultat = random.randint(1,3)
        print(f"\nTorn del jugador {jugador}:")
        pregunta, respuesta = generar_pregunta(dificultat)
        print("Pregunta:", pregunta)  # Se imprime la pregunta
        respuesta_usuario = int(input("Resposta: "))
        if respuesta_usuario == respuesta and dificultat == 1:
            mover_jugador(i, 1)  # Avanzar dos casillas si la respuesta es correcta
            print("Resposta correcte! una casella endevant.")
        elif respuesta_usuario == respuesta and dificultat == 2:
            mover_jugador(i, 2)  # Avanzar dos casillas si la respuesta es correcta
            print("Resposta correcta! dos caselles endevant.")
        elif respuesta_usuario == respuesta and dificultat == 3:
            mover_jugador(i, 3)  # Avanzar dos casillas si la respuesta es correcta
            print("Resposta correcta! tres caselles endevant.")
        elif respuesta_usuario != respuesta and dificultat == 1:
            mover_jugador(i, -1)  # Avanzar dos casillas si la respuesta es correcta
            print("Resposta incorrecte! una casella enrrere.")
        elif respuesta_usuario != respuesta and dificultat == 2:
            mover_jugador(i, -2)  # Avanzar dos casillas si la respuesta es correcta
            print("Resposta incorrecta! dos caselles enrrere.")
        elif respuesta_usuario != respuesta and dificultat == 3:
            mover_jugador(i, -3)  # Avanzar dos casillas si la respuesta es correcta
            print("Resposta incorrecta! tres caselles enrrere.")
        imprimir_tablero()
        if tauler[i, 0] >= tamany_tauler:
            meta_alcanzada = True
            print(f"\n¡El jugador {jugador} ha llegado a la meta! ¡Felicidades!")
            break
```
